```python
# Import relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in csv and get a feel of data
df = pd.read_csv("../data/car_accidents_canada.csv")
df.head()
```

```python
# Exploratory data
df.shape
```

```python
# Exploratory data
df.info()
```

```python
# Convert unknowns to NAN
df1 = df.replace(["U", "UU", "QQ", "Q", "N", "NN", "NNNN"], np.nan)
```

```python
# Convert month, hour, c_vehs and weekday to integer and drop na
df1 = df1.dropna(subset=["C_MNTH", "C_WDAY", "C_HOUR", "C_VEHS"])
df1["C_YEAR"] = df1["C_YEAR"]
df1["C_MNTH"] = df1["C_MNTH"].astype(int)
df1["C_HOUR"] = df1["C_HOUR"].astype(int)
df1["C_WDAY"] = df1["C_WDAY"].astype(int)
df1["C_VEHS"] = df1["C_VEHS"].astype(int)
print(df1.info())
```

```python
# Cleaned final data using one hot encoding
df2 = pd.get_dummies(df1, columns=["C_CONF", "C_RCFG", "C_WTHR", "C_RSUR", "C_RALN", "C_TRAF", "V_ID", "V_TYPE", "V_YEAR", "P_ID", 
                             "P_SEX", "P_AGE", "P_PSN", "P_ISEV", "P_SAFE", "P_USER"])
```

```python
# target variable
dy = pd.DataFrame()
dy["fatal_severity"] = df.C_SEV
dy.head()
```

```python
# Plot by year
df_year = df1[["C_YEAR", "C_SEV"]].groupby("C_YEAR").count()
print(df_year)

plt.bar(df_year.index, df_year.C_SEV, align="center", alpha=0.5)
plt.title('Distribution of Collision by year')
plt.show()


# Plot by year severity
dfs = df1[df1.C_SEV == 1]
dfs_year = dfs[["C_YEAR", "C_SEV"]].groupby("C_YEAR").count()
print(dfs_year)
plt.bar(dfs_year.index, dfs_year.C_SEV, align="center", alpha=0.5)
plt.title('Distribution of Severe Collision by year')
plt.show()
```

```python
# Plot by month
month = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
df_month = df1[["C_MNTH", "C_SEV"]].groupby("C_MNTH").count()
print(df_month)

plt.bar(month, df_month.C_SEV, align="center", alpha=0.5)
plt.title('Distribution of Collision by month')
plt.show()

# Plot by month severity
dfs = df1[df1.C_SEV == 1]
dfs_month = dfs[["C_MNTH", "C_SEV"]].groupby("C_MNTH").count()
print(dfs_month)
plt.bar(month, dfs_month.C_SEV, align="center", alpha=0.5)
# plt.ylabel('')
plt.title('Distribution of Severe Collision by month')
plt.show()
```

```python
# Plot by hour of day
df_hour = df1[["C_HOUR", "C_SEV"]].groupby("C_HOUR").count()
print(df_hour)

plt.plot(df_hour.index, df_hour.C_SEV, alpha=0.5)
plt.title('Distribution of Collision by hour of day')
plt.show()

# Plot by hour of day severity
dfs = df1[df1.C_SEV == 1]
dfs_hour = dfs[["C_HOUR", "C_SEV"]].groupby("C_HOUR").count()
print(dfs_hour)
plt.plot(dfs_hour.index, dfs_hour.C_SEV, alpha=0.5)
# plt.ylabel('')
plt.title('Distribution of Severe Collision by hour of day')
plt.show()
```

```python
# Plot by week day
day = ["Mon", "Tues", "Wed", "Thurs", "Frid", "Sat", "Sun"]
df_wday = df1[["C_WDAY", "C_SEV"]].groupby("C_WDAY").count()
print(df_wday)

plt.plot(day, df_wday.C_SEV, alpha=0.5)
plt.title('Distribution of Collision by weekday')
plt.show()

# Plot by week day severity
dfs = df1[df1.C_SEV == 1]
dfs_wday = dfs[["C_WDAY", "C_SEV"]].groupby("C_WDAY").count()
print(dfs_wday)
plt.plot(day, dfs_wday.C_SEV, alpha=0.5)
# plt.ylabel('')
plt.title('Distribution of Severe Collision by weekday')
plt.show()
```

```python
# Plot by weather
wea = ["Sunny", "Cloudy", "Raining", "Snowing", "Freezing", "Ltd. visib", "Windy"]
df_wthr = df1[["C_WTHR", "C_SEV"]].groupby("C_WTHR").count()
print(df_wthr)

plt.bar(wea, df_wthr.C_SEV, align="center", alpha=0.5)
plt.title('Distribution of Collision by weather condition')
plt.show()

# Plot by weather severity
dfs = df1[df1.C_SEV == 1]
dfs_wthr = dfs[["C_WTHR", "C_SEV"]].groupby("C_WTHR").count()
print(dfs_wthr)
plt.bar(wea, dfs_wthr.C_SEV, align="center", alpha=0.5)
plt.title('Distribution of Severe Collision by weather condition')
plt.show()
```

```python
# Plot by Sex
sex = ["Female", "Male"]
df_sex = df1[["P_SEX", "C_SEV"]].groupby("P_SEX").count()
print(df_sex)

explode = (0, 0.1)
fig1, ax1 = plt.subplots()
ax1.pie(df_sex.C_SEV, explode=explode, labels=sex, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.title('Distribution of Collision by sex')
plt.show()

# Plot by sex severity
dfs = df1[df1.C_SEV == 1]
dfs_sex = dfs[["P_SEX", "C_SEV"]].groupby("P_SEX").count()
print(dfs_sex)

explode = (0, 0.1)
fig1, ax1 = plt.subplots()
ax1.pie(dfs_sex.C_SEV, explode=explode, labels=sex, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.title('Distribution of Severe Collision by sex')
plt.show()
```
